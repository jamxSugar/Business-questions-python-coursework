```python
# Research on tweets text tokens
# Functions for making wordcloud graph by frequency
## Get tweets related with worldcup
def get_tweets_pages(query):
    worldcup_tweets = []
    for page in tweepy.Cursor(api.search_tweets, 
                              q=query, 
                              lang='en').pages(150):
        worldcup_tweets.append(page)
    return worldcup_tweets

## Filter emoji function
def filter_emoji(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'', text)

## remove urls function
def remove_urls (vTEXT):
    vTEXT = re.sub(r'http[s]?://\S+', '', vTEXT, flags=re.MULTILINE)
    return(vTEXT)


## Get all tweets text in list
def get_tweets_text(tweets):
    tweets_text = []
    for search_result in tweets:
        for status in search_result:  # for every tweet
            tweets_text.append(status.text)
    return tweets_text

## Filter emoji, urls, special characters, numbers and stopwrods in tweets text 
def text_filter(tweets_text):
    tweets_string = " ".join(tweets_text)
    tweets_string=remove_urls (tweets_string)
    tweets_string_filterd_emoji=filter_emoji(tweets_string)
    # Put all the tokens in a list
    tokens = word_tokenize(tweets_string_filterd_emoji)
    lowercase_tokens = [token.lower() for token in tokens]
    specal_characters=["rt","ht","r t",'``', '``', "''", '``',
                       'one','two','three','four','five','six','seven','nine','ten']
    remove_these = set(stopwords.words('english') + list(string.punctuation) + list(string.digits)+specal_characters)
    filtered_text = [token 
                     for token in lowercase_tokens 
                     if not token in remove_these]
    return filtered_text

## Create frequency dictionary for filtered_text and plot a wordcloud graph based on frequencies of each word
def get_wordcloud(filtered_text,title):
    simple_frequencies_dict = Counter(filtered_text)
    cloud = WordCloud(width=800, 
                      height=400, 
                      max_font_size=160,
                     background_color='white').generate_from_frequencies(simple_frequencies_dict)
    plt.figure(figsize=(16,12))
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title,fontsize=20)
    plt.show()
```

```python
# Get tweets from hashtags
worldcup_tweets=get_tweets_pages('ICCT20WorldCup OR T20WorldCup OR cricket')
# Get tweets_text
worldcup_tweets_text=get_tweets_text(worldcup_tweets)
# Filter tweets_text
worldcup_filtered_text=text_filter(worldcup_tweets_text)

# Save worldcup_filtered_text into file. 
#   Because the twitter data we get are real-time, the result might be changed next time we run the code.
f = open('worldcup_filtered_text.txt', "w")
f.write(' '.join(worldcup_filtered_text))
f.close()
```

```python
# Research on hashtags and cricket teams network
# Functions for extract hashtags and text in a dataFrame object, 
#    and create a network graph for all cricket teams metioned in hashtags and mentioned in text,
#    based on the frequency they appear simultaneously in one tweet.

## Create dataframe, with text and cricket teams hashtags (we get cricket teams names by GPE entities)
def create_tweets_df(tweets):
    nlp = en_core_web_sm.load()
    
    # Name the columns of df
    df=pd.DataFrame(columns=['text','hashtags'])
    for search_result in tweets:
        for status in search_result:  # for every tweet
            row_hashtags=[]
            hashtags = status.entities['hashtags']
            if len(hashtags) > 0:     # if there are hashtags
                for h in hashtags:
                    row_hashtags.append(h['text'])
            # GPE - Geo-political entity as the names of cricket teams from hashtags
            doc=nlp(' '.join(row_hashtags))
            named_entities = [(ent.text, ent.label_) for ent in doc.ents] 
            row_hashtags = [string for (string, tag) in named_entities if (tag == 'GPE') and (not string.startswith('\n'))]
            
            df=df.append({'text':status.text,'hashtags':row_hashtags}, ignore_index=True)
    return df

## Filter emoji, urls, special characters, numbers and stopwrods for df['text'] and df['hashtags']
def tweets_df_filter(df):
    for x,row in df.iterrows():
        row.text=" ".join(text_filter(row.text.split(" ")))
        row.hashtags=text_filter(row.hashtags)
    return df

## create a tsv file to save cricket teams hashtags and connected entities (cricket teams in text) 
def connections_save_to_file(header,filename):
    file_content=header
    nlp = en_core_web_sm.load()

    for x,row in df.iterrows():
        if len(row['hashtags'])>0:
            doc=nlp(row['text'])
            named_entities = [(ent.text, ent.label_) for ent in doc.ents] 
            # Choose GPE - Geo-political entity as the names of cricket teams
            countries = [string for (string, tag) in named_entities if (tag == 'GPE') and (not string.startswith('\n'))]
            if len(countries) > 0:
                for hashtag in row['hashtags']:
                    for country in countries:
                        file_content=file_content + hashtag + "\t" + country + "\n"

    f = open(filename, "w")
    f.write(file_content)
    f.close()
    
## Plot a network graph based on the connections between source and target
def plot_network(df,source,target,title):
    
    # setting the size of the network
    plt.figure(figsize=(40,40))
    # setting layout
    layout = nx.spring_layout(g, k = 0.4)

    # go through every target name, ask the graph how many
    # connections it has. Multiply that by 200 to get the circle size
    tsize = [g.degree(target_each) * 200 for target_each in target]

    # draw the target nodes
    nx.draw_networkx_nodes(g, 
                           layout, 
                           nodelist=target,
                           node_color='lightblue',
                           node_size=tsize # a list of sizes, based on g.degree
                           )

    # draw source nodes
    nx.draw_networkx_nodes(g, layout, nodelist=source, node_color='yellow', node_size=100)

    # popular target (with a degree that's greater than 2)
    popular_target = [target_each for target_each in target if g.degree(target_each) > 2 ]

    # draw popular target nodes
    nx.draw_networkx_nodes(g, layout, nodelist=popular_target, node_color='red', node_size=100)

    # draw network edges
    nx.draw_networkx_edges(g, layout, width=1, edge_color="#cccccc")

    # draw node labels
    nx.draw_networkx_labels(g, layout, font_size=12)

    plt.axis('off')
    plt.title(title, fontsize=40)
    plt.show()
```
